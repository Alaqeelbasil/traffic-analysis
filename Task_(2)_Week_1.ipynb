```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Tuwaiq/Data_Analysis/Tasks/Task 2/Task (2) Dataset.csv')
```

## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
df.describe()
```

### Identify and analyze patterns in the data.

```python

df.isnull().sum()

plt.figure(figsize=(10, 6))
sns.histplot(df['TrafficIndexLive'], kde=True)
plt.title('Distribution of Traffic Index Live')
plt.xlabel('Traffic Index Live')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='TrafficIndexLive', y='Jamscount', data=df)
plt.title('Relationship between Traffic Index Live and Jamscount')
plt.xlabel('Traffic Index Live')
plt.ylabel('Jamscount')
plt.show()

df['Datetime'] = pd.to_datetime(df['Datetime'])
df.set_index('Datetime', inplace=True)
plt.figure(figsize=(12, 6))
plt.plot(df['TrafficIndexLive'])
plt.title('Trend of Traffic Index Live over Time')
plt.xlabel('Datetime')
plt.ylabel('Traffic Index Live')
plt.show()

correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
